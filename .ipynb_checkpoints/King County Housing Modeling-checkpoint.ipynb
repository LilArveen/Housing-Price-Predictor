```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as plab
import seaborn as sns
import patsy
import statsmodels.api as sm
import scipy.stats as stats

from scipy.stats import ttest_ind, chisquare, normaltest
from sklearn.linear_model import LinearRegression


from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor

```

```python
#import data
housingData = pd.read_csv("Kings County Housing Data.csv")
housingData.head(10)
```

```python
#See total data points
housingData.describe()

```

## Exploring, Cleaning, and Visualization of the Data

```python
#see correlation between each columns/variables
correlation = housingData.corr()['price']
print(correlation)
```

Showing the correlation as a graph:

```python
clrs = ['red' if (x < 0) else 'green' for x in correlation.values]

plot = sns.barplot(correlation.index, correlation.values, palette=clrs)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```

We can see that many variables don't correlate well with price. In fact, some are negative. However, we should still keep these values, as we need to see which values are statstically significant in relation with price.

```python
x = housingData.loc[:, housingData.columns != 'price']
x = x.loc[:, x.columns!='date']
y = housingData["price"]
model = sm.OLS(y, x.astype(float)).fit()

print(model.summary())
```

```python
x = x.loc[:, x.columns!='floors']
model = sm.OLS(y, x.astype(float)).fit()

print(model.summary())
```

Let us drop the data points that aren't statstically significant

```python
housingData = housingData.drop(['date', 'floors'],axis = 1)
housingData.head(5)
```

Let us see if log transforming the data will give a clearer linear relation for the majority of variables

```python
housingLogTransform = housingData.apply(np.log)
housingLogTransform['price'] = housingLogTransform['price'].apply(np.sqrt)

```

```python
housingLogTransform.describe()
```

```python
housingLogTransform = housingLogTransform.drop(['sqft_basement', 'yr_renovated', 'long', 'waterfront', 'view'], axis = 1)
```

```python
housingLogTransform = housingLogTransform.replace([np.inf, -np.inf], np.nan)
housingLogTransform = housingLogTransform.dropna()
housingLogTransform.describe()
```

```python

#lets see which values are significantly correlated based off p value
x = housingLogTransform.loc[:, housingLogTransform.columns != 'price']
y = housingLogTransform["price"]
model = sm.OLS(y, x.astype(float)).fit()

print(model.summary())
```

```python
housingLogTransform = housingLogTransform.drop(['id', 'sqft_lot'], axis = 1)
```

Let us now check if correlation has improved through log transforming.

```python

corrLogVReg = pd.DataFrame()
correlateDF = housingData[['price','bedrooms', 'bathrooms', 'sqft_living', 'condition', 'grade', 'sqft_above', 'yr_built', 'zipcode', 'lat'
                          ,'sqft_living15', 'sqft_lot15']]
corrLogVReg['Change'] = (housingLogTransform.corr()['price']-correlateDF.corr()['price'])
clrs = ['red' if (x < 0) else 'green' for x in corrLogVReg['Change']]

plot = sns.barplot(corrLogVReg.index, corrLogVReg.Change, palette=clrs)
plot.set_xticklabels(plot.get_xticklabels(), rotation=30)

print("Average Improvement in correlation: " )
print(corrLogVReg['Change'].mean())
```

We can see that while some variables improved, other values decreased in terms of correlation. Some, such as waterfront, and views, were rendered unusable due to the log transformation.

Not only that, but the variables that had increased correlation didn't increase very much. This can be seen if we compare the original values against price vs the log transformed values against price:

```python
sns.pairplot(y_vars = ['price'], x_vars =['bedrooms', 'condition', 'grade', 'yr_built', 'zipcode', 'lat'
                          ,'sqft_living15', 'sqft_lot15'], data = correlateDF)
```

```python
sns.pairplot(y_vars = ['price'], x_vars =['bedrooms', 'condition', 'grade', 'yr_built', 'zipcode', 'lat'
                          ,'sqft_living15', 'sqft_lot15'], data = housingLogTransform)
```

While some of the log transformed values have a clearer linear relation, such as sqft_living15, the vast majority, such as zipcode and lat still have little to no clear improvement in terms of linear relation. We will thus continue to utilize the original data instead of the log transformed data.

## Modeling And Prediction

```python
price = housingData['price']
training = housingData.drop(['price'], axis = 1)

x_train, x_test, y_train, y_test = train_test_split(training, price, test_size = 0.1, random_state = 2)


```

```python
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import r2_score
```

### Linear Regression

```python
linReg = LinearRegression()

linReg.fit(x_train, y_train)
y_linearPred = linReg.predict(x_test)
linearModel = pd.DataFrame()
linearModel['Predicted'] = y_linearPred
linearModel['Actual'] = y_test.values


plt.xticks(rotation=45)
sns.regplot(x=linearModel['Actual'], y=linearModel['Predicted'])
         
```

### Assesment and Accuracy of Linear Model

Checking R-Squared value

```python
r2_score(y_test, y_linearPred, multioutput='variance_weighted')
```

## Gradient Boosting Regression

```python
gbr = GradientBoostingRegressor(n_estimators = 500, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.05, loss = 'ls')
gbr.fit(x_train, y_train)
y_pred = gbr.predict(x_test)

gbrModel = pd.DataFrame()
gbrModel['Predicted'] = y_pred
gbrModel['Actual'] = y_test.values


plt.xticks(rotation=45)
sns.regplot(x=gbrModel['Actual'], y=gbrModel['Predicted'])

```

```python
r2_score(y_test, y_pred, multioutput='variance_weighted')
```
